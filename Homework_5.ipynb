### 1. Write the keras code for a multilayer fully connected neural network with the following structure: Three hidden layers. 50 hidden units in the first hidden layer, 100 in the second, and 150 in the third. Activate all hidden layers with relu. The output layer should be built to classify to five categories. Your optimization technique should be adam and you should score using ROC AUC. You can assume the input data has 100 features.

```python
# Load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from keras.layers import Dense, Activation, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import SGD
from keras import backend as K

# Multilayer fully connected neural network - 5-class classification
# 3 hidden layers: 50, 100, 150 units (all ReLU)
# Output: 5 categories (softmax)
# Input: 100 features

model_1 = Sequential([
    Dense(50, input_shape=(100,)),
    Activation('relu'),
    Dense(100),
    Activation('relu'),
    Dense(150),
    Activation('relu'),
    Dense(5),
    Activation('softmax'),])

model_1.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['auc'])

model_1.summary()
```

### 2. Write the keras code for a multilayer fully connected neural network with the following structure: Two hidden layers. 75 hidden units in the first hidden layer and 150 in the second. Activate all hidden layers with relu. The output layer should be built to classify a binary dependent variable. Your optimization technique should be adam and you should score using ROC AUC. You can assume the input data has 100 features.


```python
# Multilayer fully connected neural network - binary classification
# 2 hidden layers: 75, 150 units (all ReLU)
# Output: binary (sigmoid, 1 unit)
# Input: 100 features

model_2 = Sequential([
    Dense(75, input_shape=(100,)),
    Activation('relu'),
    Dense(150),
    Activation('relu'),
    Dense(1),
    Activation('sigmoid'),])

model_2.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['auc'])

model_2.summary()
```

### 3. Write the keras code for a convolutional neural network with the following structure: Two convolutional layers with a 3x3 filter size, with 16 filters in the first layer and 28 in the second. Activate all convolutional layers with relu. Use max pooling after each convolutional layer with a 2x2 filter. The output layer should be built to classify to ten categories. Your optimization technique should be adam and you should score using ROC AUC. You can assume the input data has a shape of 10x10.

```python
# CNN - 10-class classification
# 2 Conv2D layers (3x3 filter): 16 filters, then 28 filters (both ReLU)
# MaxPooling (2x2) after each conv layer
# Output: 10 categories (softmax)
# Input shape: 10x10 (1 channel assumed)

model_3 = Sequential()

# First convolutional layer
model_3.add(Conv2D(16, (3, 3), padding='same', input_shape=(10, 10, 1)))
model_3.add(Activation('relu'))
model_3.add(MaxPooling2D(pool_size=(2, 2)))

# Second convolutional layer
model_3.add(Conv2D(28, (3, 3), padding='same'))
model_3.add(Activation('relu'))
model_3.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten and output layer
model_3.add(Flatten())
model_3.add(Dense(10))
model_3.add(Activation('softmax'))

model_3.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['auc'])

model_3.summary()
```

### 4. Write the keras code for a convolutional neural network with the following structure: Two convolutional layers with a 3x3 filter size, with 32 filters in the first layer and 32 in the second. Activate all convolutional layers with relu. Use max pooling after each convolutional layer with a 2x2 filter. Add two fully connected layers with 128 hidden units in each layer and relu activations. The output layer should be built to classify to six categories. Your optimization technique should be adam and you should score using ROC AUC. You can assume the input data has a shape of 10x10.

```python
# CNN - 6-class classification
# 2 Conv2D layers (3x3 filter): 32 filters, then 32 filters (both ReLU)
# MaxPooling (2x2) after each conv layer
# 2 fully connected layers with 128 hidden units (ReLU)
# Output: 6 categories (softmax)
# Input shape: 10x10 (1 channel assumed)

model_4 = Sequential()

# First convolutional layer
model_4.add(Conv2D(32, (3, 3), padding='same', input_shape=(10, 10, 1)))
model_4.add(Activation('relu'))
model_4.add(MaxPooling2D(pool_size=(2, 2)))

# Second convolutional layer
model_4.add(Conv2D(32, (3, 3), padding='same'))
model_4.add(Activation('relu'))
model_4.add(MaxPooling2D(pool_size=(2, 2)))

# Fully connected layers
model_4.add(Flatten())
model_4.add(Dense(128))
model_4.add(Activation('relu'))
model_4.add(Dense(128))
model_4.add(Activation('relu'))

# Output layer
model_4.add(Dense(6))
model_4.add(Activation('softmax'))

model_4.compile(loss='categorical_crossentropy',
                optimizer='adam',
                metrics=['auc'])

model_4.summary()
```

### 1. Describe the main differences between supervised and unsupervised learning.


In supervised learning, we observe both a set of features X and a response variable Y, and the goal is to predict Y using X. In unsupervised learning, we observe only the features X with no associated response variable Y, and the goal is instead to discover interesting structure or subgroups within the data. Supervised learning is more objective because performance can be measured against known outcomes, while unsupervised learning is more subjective since there is no gold standard to evaluate against. One practical advantage of unsupervised learning is that unlabeled data is often much easier to obtain than labeled data, which typically requires human intervention.

### 2. Which supervised learning approaches did we cover this semester? What are the major differences between these techniques?

The supervised learning approaches we covered this semester include:
- K-Nearest Neighbors (KNN) — an instance-based model that classifies new observations based on the majority class of their K nearest neighbors in the training data. No explicit model is learned; predictions are made directly from the training data.
- Linear/Logistic Regression (including Ridge & Lasso) — parametric linear models. Linear regression predicts continuous outcomes, logistic regression predicts categorical outcomes. Ridge and Lasso add regularization penalties to reduce overfitting.
- Support Vector Machines (SVMs) — a kernel/margin-based model that finds the maximum-margin decision boundary separating classes. Kernels allow it to handle non-linear boundaries.
- Decision Trees — a tree-based model that recursively splits data on feature thresholds. Highly interpretable but prone to overfitting.
Random Forests — an ensemble of decision trees trained on random bootstrap samples with random feature subsets. Reduces the variance problem of individual trees.
- Gradient Boosting — another tree-based ensemble method, but trees are built sequentially, each correcting the residual errors of the previous one. Often achieves very high accuracy.
- Neural Networks — multi-layer networks of weighted connections and activation functions, capable of learning highly complex non-linear patterns but the least interpretable.

The major differences across these techniques come down to a few key axes: linearity (regression models assume linear boundaries; the rest do not), interpretability (regression and decision trees are most transparent; neural networks are black-box), how predictions are made (KNN uses stored training data directly; parametric models learn explicit parameters; tree models learn splitting rules), data requirements (neural networks need the most data; KNN and logistic regression can work with less), and computational cost (KNN is cheap to train but slow to predict; neural networks are expensive both ways).

### 4. Which unsupervised learning modeling approaches did we cover this semester? What are the major differences between these techniques?

The unsupervised learning approaches covered this semester include:

- Principal Components Analysis (PCA) — a dimensionality reduction technique that finds linear combinations of the features (called principal components) that capture maximum variance, and are mutually uncorrelated. It does not assign observations to groups but instead produces a continuous low-dimensional representation of the data. It is also useful for data visualization by projecting high-dimensional data onto 2–3 components that can be plotted.
- K-Means Clustering — partitions observations into a pre-specified number of clusters K by iteratively assigning each observation to its nearest centroid and recomputing centroids until assignments stop changing. Requires specifying K in advance and assumes clusters are roughly spherical and similarly sized.
- Hierarchical (Agglomerative) Clustering — builds a tree-like structure called a dendrogram by successively merging the two most similar clusters, starting from each observation in its own cluster. Does not require specifying K in advance — the dendrogram can be cut at any height to yield any number of clusters.
- Manifold Learning — a non-linear dimensionality reduction technique (covered in Class 13) that, unlike PCA, can capture complex curved structures in high-dimensional data that linear methods would miss.

The major differences across these techniques come down to a few key axes: goal (PCA and manifold learning reduce dimensionality; clustering finds discrete subgroups), linearity (PCA is a linear method; manifold learning is non-linear), output (PCA and manifold learning produce continuous coordinates; K-Means and hierarchical clustering produce discrete cluster assignments), and need to specify K (K-Means requires K up front; hierarchical clustering does not).

### 7. With regard to neural networks, what are 3 major differences between a fully connected deep neural network and a convolutional neural network model? Be sure to define any key terms in your explanation.

1. **Local vs. global connectivity**: In a fully connected, dense network, every neuron in one layer is connected to every neuron in the next layer. Each neuron sees the entire input. In a CNN, neurons in a convolutional layer are each connected only to a small local patch of the input, defined by the kernel (filter) size (e.g., 3x3 pixels). This local connectivity allows CNNs to detect spatial features like edges and textures that are inherently local in nature.
2. **Weight sharing**: In a CNN, the same filter weights are shared across all spatial positions as the filter slides over the input, a property called weight sharing. This dramatically reduces the number of parameters compared to a fully connected network applied to the same input, making CNNs far more efficient for image data. In a fully connected network, every connection has its own unique weight, so the parameter count grows rapidly with input size.
3. **Pooling layers for spatial invariance**: CNNs include pooling layers (e.g., MaxPooling) that downsample feature maps by taking the maximum value in each local region. This introduces translational invariance since the model becomes less sensitive to the exact position of a detected feature within the input. Fully connected networks have no equivalent mechanism and treat the position of every input value as fixed and equally distinct.



### 8. Which models did we learn about this semester that easily offer meaningful information about which features are more relevant? For instance, if I have a model predicting sales revenue or purchase activity based on different features, which models can clearly tell me which features are most important for driving that behavior?

The models covered this semester that most easily offer meaningful information about which features are more relevant include:

- **Linear/Logistic Regression** (including Ridge & Lasso): the model coefficients directly encode each feature's contribution to the predicted outcome. A larger coefficient magnitude means a stronger effect. Ridge and Lasso are especially useful here because Lasso can shrink irrelevant feature coefficients all the way to zero, effectively selecting the most important features automatically.
- **Decision Trees**: the tree structure is fully transparent. You can see exactly which features are used at each split and how much each split reduces impurity (e.g., Gini or entropy). Feature importance scores are straightforward to extract and interpret.
- **Random Forests**: aggregates feature importances across all trees in the ensemble via mean decrease in impurity or permutation importance, producing a robust ranked list of which features most drive predictions.
- **Gradient Boosting**: similarly provides importance scores based on how frequently and how much each feature contributes to splits across all boosted trees.

In contrast, the models covered this semester that do not easily offer feature importance are:

- KNN: makes predictions based on distance to neighbors with no learned parameters, so there are no coefficients or splits to inspect.
- SVMs: with non-linear kernels, the decision boundary is not directly interpretable in terms of individual features.
- Neural Networks / CNNs — black-box models with no native feature importance output.

So for a use case like predicting sales revenue or purchase activity, **Logistic Regression, Decision Trees, Random Forests, and Gradient Boosting** would be the models to turn to if understanding which features drive the outcome is a priority.
