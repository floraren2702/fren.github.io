# Project 1

This project explores class sizes in New York City public schools during the 2017вЂ“2018 academic year.
The dataset reports, at a classroom level, how many students were enrolled across schools and grades.

This dataset contains thousands of rows (each representing a group of students in a school), including numeric fields such as Number of Students, which I focus on for this project.


This notebook walks through a simple analysis of class sizes using two approaches:

1. With pandas (the easy way)
2. With only the Python standard library (the вЂњhard wayвЂќ)

It concludes with a simple data visualization to show how class sizes vary.


Data source: 
NYC Open Data вЂ“ 2017вЂ“2018 School Level Class Size Report

https://data.cityofnewyork.us/Education/2017-2018-SCHOOL-LEVEL-CLASS-SIZE-REPORT/2ic6-4hrf/about_data




```python
## 1. Easy way (Pandas)

import pandas as pd

# ------------------------------------------------------------
# 1) Load the dataset
# ------------------------------------------------------------
# Reads the CSV file into a pandas DataFrame.
df = pd.read_csv("2017-2018_SCHOOL_LEVEL_CLASS_SIZE_REPORT_20251107.csv")
print("Number of rows:", len(df))

# We focus on the column "Number of Students".
students = df["Number of Students"]

# ------------------------------------------------------------
# 2) Compute summary statistics
# ------------------------------------------------------------

# Mean: the average number of students
mean_val = students.mean()

# Median: the middle value when sorted
median_val = students.median()

# Mode: the most frequently occurring value
# .mode() returns a Series, so we select the first element
mode_val = students.mode()[0]

# ------------------------------------------------------------
# 3) Print results
# ------------------------------------------------------------
print("Mean number of students:", round(mean_val, 2))
print("Median number of students:", round(median_val, 2))
print("Mode number of students:", round(mode_val, 2))
```

```python
## 2. Hard way (Standard Library Only)

import csv

# ------------------------------------------------------------
# 1) Load the dataset
# ------------------------------------------------------------
filename = "2017-2018_SCHOOL_LEVEL_CLASS_SIZE_REPORT_20251107.csv"

# We focus on the column "Number of Students".
col = "Number of Students"

student_counts = []

# Reads data from the CSV
with open(filename, newline="") as cvsfile:
    # DictReader allows us to look up values by column name.
    reader = csv.DictReader(cvsfile)
    # Collect all numeric values from the column
    for row in reader:
        # Since values are always valid, convert directly
        student_counts.append(float(row[col]))

# ------------------------------------------------------------
# 2) Compute summary statistics
# ------------------------------------------------------------

# Mean: the average number of students
mean_students = sum(student_counts) / len(student_counts)

# Median: the middle value when sorted
sorted_counts = sorted(student_counts)
n = len(sorted_counts)
mid_index = n // 2

if n % 2 == 1:
    # Odd number of entries: middle value
    median_students = sorted_counts[n // 2]
else:
    # Even number of entries: average of the two middle values
    median_students = (sorted_counts[n // 2 - 1] + sorted_counts[n // 2]) / 2

# Mode: the most frequently occurring value
# Use a dictionary to count occurrences
frequency = {}

for value in student_counts:
    if value in frequency:
        frequency[value] += 1
    else:
        frequency[value] = 1

# Find the mode by checking the dictionary as we go
mode_value = None
mode_count = 0

for value, count in frequency.items():
    if count > mode_count:
        mode_count = count
        mode_value = value

# ------------------------------------------------------------
# 3) Print results
# ------------------------------------------------------------
print(f"Mean ({col}):   {mean_students:.2f}")
print(f"Median ({col}): {median_students:.2f}")
print(f"Mode ({col}):   {mode_value:.2f}")
```

```python
# ------------------------------------------------------------
# Data Visualization
# ------------------------------------------------------------

data = student_counts

print("\nClass Size Visualization (1 emoji per 10 students):")
print("(showing first 200 rows)\n")

# Loop through each value in the data
# enumerate() gives us both i, the row number (1, 2, 3, вЂ¦) and value, the actual class size
for i, value in enumerate(data[:200], start=1):
    # Determine how many emojis to draw for this row
    # We use integer division (//) to group into bins of size 10.
    # Examples: 0вЂ“9 students: 0 emojis, 10вЂ“19 students: 1 emoji, 20вЂ“29 students: 2 emojis
    bar_count = int(value) // 10

    # Create the visual bar using repeated emoji symbols
    bar = "рџ‘©вЂЌрџЏ«" * bar_count

    print(f"{i:>3}. {int(value):>4} | {bar}")
```

# Key Takeaways

NYC class sizes show a fairly concentrated central range, most classrooms fall within a predictable band.
The median is a helpful summary because class sizes are not perfectly symmetric; a few especially large or small classes can shift the mean.
The mode reinforces this: certain class sizes are especially common.
