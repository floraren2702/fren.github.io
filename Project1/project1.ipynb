```python
import pandas as pd

df = pd.read_csv("2017-2018_SCHOOL_LEVEL_CLASS_SIZE_REPORT_20251107.csv")
df

df = df["Number of Students"]

# Compute the mean
df.mean()

# Compute the median
df.median()

# Compute the mode
df.mode()

print(df.mean())
print(df.median())
print(df.mode())
```

```python
import csv

filename = "2017-2018_SCHOOL_LEVEL_CLASS_SIZE_REPORT_20251107.csv"
col = "Number of Students"

values = []

with open(filename, newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        v = row[col]
        values.append(float(v))
print(values)

mean_h = sum(values) / len(values)

values_sorted = sorted(values)
n = len(values_sorted)

if n % 2 == 1:
    median_h = values_sorted[n // 2]
else:
    median_h = (values_sorted[n // 2 - 1] + values_sorted[n // 2]) / 2

counts = {}
for v in values:
    counts[v] = counts.get(v, 0) + 1

mode_h = max(counts, key=counts.get)

print(mean_h)
print(median_h)
print(mode_h)
```

```python
# Scale so bars are not huge
import numpy as np

print(np.min(values))
print(np.max(values))


max_len = 500
max_val = max(values)

for v in values:
    count = int((v / max_val) * max_len)
    print(f"{v:4.0f} | {'👤' * count}")
```
